## - find의 다양한 문법 - nor
* not or

```python
docs = actor.find({'$nor':[{'흥행지수':{'$gte':10000}},{'흥행지수':{'$lte':7000}}]},{'흥행지수':1,'랭킹':1,'배우이름':1,'_id':0}).sort('랭킹')
for doc in docs:
    print(doc)
```

## - find의 다양한 문법 - in, nin
* in: 들어가 있다.
* nin: not in - 들어가 있지 않다.

```python
docs = actor.find({'흥행지수':{'$in':[39995,9102]}},{'흥행지수':1,'랭킹':1,'배우이름':1,'_id':0}).sort('랭킹')
for doc in docs:
    print(doc)
```

```python
docs = actor.find({'배우이름':{'$in':['조인성']}},{'_id':0})
for doc in docs:
    print(doc)
```

```python
docs = actor.find({'흥행지수':{'$nin':[39995,9102]}},{'흥행지수':1,'랭킹':1,'배우이름':1,'_id':0}).sort('랭킹').limit(3)
for doc in docs:
    print(doc)
```

#### 흥행지수 가 8706, 8565가 아니고, 10000 이하인 데이터를 3개만 검색하세요. (nor, in, gt 활용, 배우이름과 흥행지수만 출력)

```python
docs = actor.find({'$nor':[{'흥행지수':{'$in':[8706,8565]}},{'흥행지수':{'$gt':10000}}]},{'배우이름':1,'흥행지수':1,'_id':0}).limit(3)
for doc in docs:
    print(doc)
```

## - find의 다양한 문법 (elemMatch)
- 적어도 한 개 이상의 리스트 요소가 복수 개의 조건을 동시에 만족하는 경우

```python
from bs4 import BeautifulSoup
import requests
import pymongo
import re

conn = pymongo.MongoClient()
actor_db = conn.cine21
elemmatch_sample = actor_db.sample
```

```python
elemmatch_sample.insert_many([
    {'results': [82, 85, 88]},
    {'results': [75, 88, 91]}
])
```

```python
docs = elemmatch_sample.find({'results': {'$gte': 90, '$lt':80}})
for doc in docs:
    print (doc)
```

```python
docs = elemmatch_sample.find({'results': {'$gte': 90, '$lt':85}})
for doc in docs:
    print (doc)
```

```python
docs = elemmatch_sample.find({'results': {'$elemMatch': {'$gte':75, '$lt':80}}})
for doc in docs:
    print (doc)
```
